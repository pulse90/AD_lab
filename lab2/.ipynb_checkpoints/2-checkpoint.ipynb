```python
import numpy as np


values = np.array([0, 1, 2, 3, 4, 5])
freqs  = np.array([27, 96, 58, 54, 18, 7])

N = freqs.sum()

# Mean
mean_val = np.sum(values * freqs) / N

# Median (expand data using frequencies)
data = np.repeat(values, freqs)
median_val = np.median(data)

# Mode (value with highest frequency)
mode_val = values[np.argmax(freqs)]

print("Total N:", N)
print("Mean   :", mean_val)
print("Median :", median_val)
print("Mode   :", mode_val)
```

```python
import numpy as np


intervals = [
    (59.95, 61.95),
    (61.95, 63.95),
    (63.95, 65.95),
    (65.95, 67.95),
    (67.95, 69.95),
    (69.95, 71.95),
    (71.95, 73.95),
    (73.95, 75.95),
]


freqs2 = np.array([5, 3, 15, 40, 17, 12, 7, 1])
N2 = freqs2.sum()

# (a) percentage of heights < 65.95 inches
count_less_65_95 = freqs2[0] + freqs2[1] + freqs2[2]
perc_less_65_95  = count_less_65_95 / N2 * 100

# (b) percentage of heights between 61.95 and 65.95 inches
count_between_61_95_65_95 = freqs2[1] + freqs2[2]
perc_between_61_95_65_95  = count_between_61_95_65_95 / N2 * 100

# (c) number of players between 61.95 and 71.95 inches
count_between_61_95_71_95 = freqs2[1] + freqs2[2] + freqs2[3] + freqs2[4] + freqs2[5]

print("Total players:", N2)
print("(a) < 65.95 in  :", count_less_65_95, "players ->", perc_less_65_95, "%")
print("(b) 61.95–65.95 :", count_between_61_95_65_95, "players ->", perc_between_61_95_65_95, "%")
print("(c) 61.95–71.95 :", count_between_61_95_71_95, "players")
```

```python
import numpy as np

value = np.array([78, 82, 85, 90, 92, 94, 95, 100, 110, 120, 150])

Q1 = np.percentile(value, 25)
Q2 = np.percentile(value,50)
Q3 = np.percentile(value, 75)
IQR = Q3 - Q1

print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)
print("IQR =", IQR)
```
