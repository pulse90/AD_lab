```python
import numpy as np


values = np.array([0, 1, 2, 3, 4, 5])
freqs  = np.array([27, 96, 58, 54, 18, 7])

N = freqs.sum()

# Mean
mean_val = np.sum(values * freqs) / N

# Median (expand data using frequencies)
data = np.repeat(values, freqs)
median_val = np.median(data)

# Mode (value with highest frequency)
mode_val = values[np.argmax(freqs)]

print("Total N:", N)
print("Mean   :", mean_val)
print("Median :", median_val)
print("Mode   :", mode_val)
```

```python
value = [78, 82, 85, 90, 92, 94, 95, 100, 110, 120, 150]
value.sort()

n = len(value)
Q2 = value[n//2]


lower_half = value[:n//2]
upper_half = value[n//2+1:]

Q1 = lower_half[len(lower_half)//2]
Q3 = upper_half[len(upper_half)//2]

IQR = Q3 - Q1
lf = Q1 - 1.5 * IQR
uf = Q3 + 1.5 * IQR
outliers = [x for x in value if x < lf or x > uf]
print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)
print("IQR =", IQR)
print("lower fence =", lf)
print("upper fence =", uf)
print("Outliers =", outliers)
```

```python
import numpy as np


intervals = [
    (2,4),
    (4,6),
    (6,8),
    (8,10)
   
]


freqs2 = np.array([3,4,2,1])
N2 = freqs2.sum()
midpoints = np.array([(a + b) / 2 for a, b in intervals])
mean = np.sum(midpoints * freqs2) / N2

print("Midpoints:", midpoints)
print("Total frequency (N):", N2)
print("Mean =", mean)
variance = np.sum(freqs2 * (midpoints - mean)**2) / N2
std_dev = np.sqrt(variance)
print("Mean =", mean)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```
