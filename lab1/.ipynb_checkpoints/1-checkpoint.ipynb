```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Bitcoin dataset (raw string for path)
bitcoin = pd.read_csv(r'C:\Users\KIIT0001\Downloads\BTC-2017min.csv.zip')

print("Sample bitcoin dataset:")
print(bitcoin.head())

# Plot histogram of 'high' prices
plt.figure(figsize=(8, 6))
plt.hist(bitcoin['high'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Bitcoin High Prices')
plt.xlabel('High Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='high', y='low', hue='close', data=bitcoin, palette='viridis')
plt.title('Scatter Plot of Bitcoin High vs Low')
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.show()
```

```python
numeric_cols = bitcoin.select_dtypes(include='number')

# Compute correlation matrix
correlation_matrix = numeric_cols.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Bitcoin Dataset')
plt.show()
```

```python
numeric_cols = bitcoin.select_dtypes(include='number')


correlation_matrix = numeric_cols.corr()


correlations_with_close = correlation_matrix['close'].drop('close')
abs_correlations = correlations_with_close.abs()


colors = ['#FF9999','#66B2FF','#99FF99','#FFCC99','#C2C2F0','#FFB266','#FF66B2']

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(abs_correlations, labels=abs_correlations.index, autopct='%1.1f%%',
        startangle=140, colors=colors[:len(abs_correlations)])
plt.title('Absolute Correlation of Features with Bitcoin Close Price')
plt.show()
```
